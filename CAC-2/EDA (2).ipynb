```python
#Importing the required libraries for EDA

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data into the data frame

df = pd.read_csv("Blank QuizCAC2.csv")
# To display the top 5 rows 
df.head(5) 
```

```python
#to display bottom 5 rows

df.tail(5)
```

```python
#checking the types of data

df.dtypes
```

```python
 #Dropping irrelevant columns

df = df.drop(['Timestamp', 'Date and Time', 'Email (CHRIST ID)', 'NAME'], axis=1)
df.head(5)
```

```python
#Renaming the columns

df = df.rename(columns={"GENDER\n": "Gender", "DEPARTMENT\n": "Department" })
df.head(5)

```

```python
#Dropping the duplicate rows

df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
# Used to count the number of rows

df.count()      
```

```python
#removing duplicates

df = df.drop_duplicates()
df.head(5)
```

```python
#counting numbe of rows after removing the duplicates

df.count()
```

```python
df['Gender'].value_counts()
```

```python
df['Department'].value_counts()
print(df.isnull().sum())
```

```python
# creating graphs for our responses

department_counts = df['Department'].value_counts()

# Creating a pie chart

plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Department Vs Responses')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

plt.show()
```
